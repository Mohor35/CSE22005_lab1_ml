```python
import pandas as pd
import numpy as np

def load_data(file_path):


    try:
        return pd.read_excel('Purchase Data.xlsx')
    except FileNotFoundError:
        print(f"Error: Excel file ('Purchase Data.xlsx') not found.")
        return None

def extract_product_data(data):


    product_quantities = data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values
    payments = data['Payment (Rs)'].values.reshape(-1, 1)
    return product_quantities, payments

def calculate_product_costs(product_quantities, payments):
    """Calculates the cost of each product using the pseudo-inverse method.

    Args:
        product_quantities (np.ndarray): Matrix of product quantities.
        payments (np.ndarray): Vector of payments.

    Returns:
        np.ndarray: Array containing the cost of each product, or None if
            an error occurs during calculation.
    """

    try:
        pseudo_inverse_A = np.linalg.pinv(product_quantities)
        product_costs = pseudo_inverse_A @ payments
        return product_costs
    except (np.linalg.LinAlgError, ValueError) as e:
        print(f"Error: Calculation failed due to {type(e).__name__}.")
        return None

def print_results(product_costs, rank):
    """Prints the results of the analysis.

    Args:
        product_costs (np.ndarray): Array containing the cost of each product,
            or None if error.
        rank (int): Rank of the matrix of product quantities.
    """

    if product_costs is not None:
        print(f"Rank of the matrix of product quantities: {rank}")
        print("Cost of each product:")
        print(f"  - Candies: ₹{product_costs[0][0]:.2f} per unit")
        print(f"  - Mangoes: ₹{product_costs[1][0]:.2f} per kg")
        print(f"  - Milk Packets: ₹{product_costs[2][0]:.2f} per packet")
    else:
        print("Error: Product cost calculation failed.")

def main():
    """Main execution of the script."""

    # Replace with the actual path to your Excel file
    file_path = "Purchase Data.xlsx"

    # Load data
    data = load_data('Purchase Data.xlsx')

    if data is not None:
        # Extract product data
        product_quantities, payments = extract_product_data(data)

        # Calculate product costs and rank
        product_costs = calculate_product_costs(product_quantities, payments)
        rank = np.linalg.matrix_rank(product_quantities)  # Calculate rank

        # Print results
        print_results(product_costs, rank)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics

def analyze_stock_data('IRCTC stock price.xlsx'):
    """
    Analyzes stock price data from an Excel file.
    """

    Args:
        file_path (str): Path to the Excel file.
    """

    # Load the data
    df = pd.read_excel('IRCTC stock price.xlsx')

    # Calculate mean and variance of price
    price_mean = statistics.mean(df['Price'])
    price_variance = statistics.variance(df['Price'])
    print("Mean price:", price_mean)
    print("Variance of price:", price_variance)

    # Select price data for Wednesdays
    wednesday_prices = df[df['Day'] == 'Wed']['Price']
    wednesday_mean = statistics.mean(wednesday_prices)
    print("Mean price on Wednesdays:", wednesday_mean)
    print("Comparison with population mean:", wednesday_mean - price_mean)

    # Select price data for April
    april_prices = df[df['Date'].dt.month == 4]['Price']
    april_mean = statistics.mean(april_prices)
    print("Mean price in April:", april_mean)
    print("Comparison with population mean:", april_mean - price_mean)

    # Probability of making a loss
    loss_probability = len(df[df['Chg%'] < 0]) / len(df)
    print("Probability of making a loss:", loss_probability)

    # Probability of making a profit on Wednesday
    wednesday_profit_probability = len(df[(df['Day'] == 'Wed') & (df['Chg%'] > 0)]) / len(df[df['Day'] == 'Wed'])
    print("Probability of making a profit on Wednesday:", wednesday_profit_probability)

    # Conditional probability of making profit given it's Wednesday
    conditional_profit_probability = len(df[(df['Day'] == 'Wed') & (df['Chg%'] > 0)]) / len(df[df['Chg%'] > 0])
    print("Conditional probability of making profit given it's Wednesday:", conditional_profit_probability)

    # Scatter plot of Chg% against day of the week
    df.plot.scatter(x='Day', y='Chg%')
    plt.title("Chg% vs Day of the Week")
    plt.show()

# Replace 'your_file.xlsx' with the actual file path
file_path = 'IRCTC stock price.xlsx'
analyze_stock_data('IRCTC stock price.xlsx')
```
